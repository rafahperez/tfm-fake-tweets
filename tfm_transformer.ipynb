```
import random
seed = random.randint(0, 2**12-1)

# Number og tweets to generate
number_tweets = 10
seed
```

```
val = !python transformers/examples/text-generation/run_generation.py \
--model_type gpt2 \
--model_name_or_path output/US_tweets_cleaned_2/ \
--length 100 \
--prompt "<BOS>" \
--stop_token "<EOS>" \
--seed $seed \
--k 100 \
--repetition_penalty 1 \
--temperature 0.7 \
--num_return_sequences $number_tweets
```

```
val.get_list()
```

```
generated_tweets = [["Tweet "+str(k+1), val[-1-2*k]] for k in range(number_tweets)[::-1]]
```

```
# retrieve last run
project = %env WANDB_PROJECT
wandb_id = wandb.api.list_runs(project)[0]['name']
```

```
# Log results on our previous wandb run
wandb.init(id=wandb_id, resume='must')
wandb.log({'examples': wandb.Table(data=generated_tweets, columns=["Tweet number","Text"])})
```

```
import pandas as pd
pd.DataFrame(generated_tweets, columns=["Tweet Number", "Text"]).to_csv("output/US_tweets_cleaned_2/generated_tweets/all_tweets_06_09_22_22_tempe_0_7.csv")
```
